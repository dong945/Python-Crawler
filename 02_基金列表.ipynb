```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import re
import pickle

# txtYEAR = "2019"
# txtMONTH = "01"
# txtGROUPID = "EUCA000500"
# url="https://www.sitca.org.tw/ROC/Industry/IN2422.aspx?txtYEAR=2019&txtMONTH=01&txtGROUPID=EUCA000500"
# url="https://www.sitca.org.tw/ROC/Industry/IN2422.aspx?txtYEAR="+txtYEAR+"&txtMONTH="+txtMONTH+"&txtGROUPID="+txtGROUPID
url_params = {'txtYEAR':'2019', 'txtMONTH':'01'}
url_headers = {"user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/72.0.3626.119 Safari/537.36"}
url = "https://www.sitca.org.tw/ROC/Industry/IN2421.aspx"
r = requests.get(url, headers = url_headers, params = url_params)
r.encoding = "utf8"
print(r)
soup = BeautifulSoup(r.text, "lxml")
Trs = soup.find_all("tr", attrs={"class":["DTeven","DTodd"]})
```

```python
pattern = "EUCA\w+"
df = []
columnName = ['基金代號','基金名稱','1個月','3個月','6個月','1年','2年','3年','5年','累計','十年','年化標準差','Sharpe']
for tr in Trs:
    tds = tr.find_all("td")
    tmp=[]
    for idx in range(len(tds)):
        if(idx == 0):
            tmp.append(re.findall(pattern, tds[idx].a["href"])[0])
        if(tds[idx].string.strip()==''):
            tmp.append(np.nan)
        else:
            tmp.append(tds[idx].string.strip())
    df.append(tmp)
df = pd.DataFrame(df, columns = columnName)
df.head(10)
```

```python
fo = open('fundList.pickle', 'wb')
pickle.dump(df, fo)
fo.close()
```
